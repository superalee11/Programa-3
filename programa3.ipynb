```python
import math

def leermatriz(n):
    print("Introduce los coeficientes de la matriz:")
    matriz = []
    vector = []
    for i in range(n):
        fila = []
        for j in range(n):
            while True:
                try:
                    elemento = float(input(f"Elemento [{i+1}][{j+1}]: "))
                    fila.append(elemento)
                    break
                except ValueError:
                    print("Entrada invalida. Intente de nuevo.")
        matriz.append(fila)
        while True:
            try:
                valor = float(input(f"Introduce el valor del vector independiente en [{i+1}]: "))
                vector.append(valor)
                break
            except ValueError:
                print("Entrada invalida. Intente de nuevo.")
    return matriz, vector

def mostrarmatriz(n, matriz, vector):
    print("\nMatriz:")
    for i in range(n):
        print(" ".join(f"{matriz[i][j]:8.3f}" for j in range(n)), f"| {vector[i]:8.3f}")

def corregircoeficiente(n, matriz, vector):
    while True:
        try:
            fila = int(input("Fila a corregir [i]: ")) - 1
            if fila < 0 or fila >= n:
                raise ValueError
            columna = int(input("Columna a corregir [j] (o 0 para el vector independiente): "))
            if columna < 0 or columna > n:
                raise ValueError
            break
        except ValueError:
            print("Entrada invalida. Intente de nuevo.")

    if columna == 0:
        while True:
            try:
                vector[fila] = float(input(f"Introduce el nuevo valor del vector independiente en [{fila+1}]: "))
                break
            except ValueError:
                print("Entrada invalida. Intente de nuevo.")
    else:
        while True:
            try:
                matriz[fila][columna-1] = float(input(f"Introduce el nuevo valor para [{fila+1}][{columna}]: "))
                break
            except ValueError:
                print("Entrada invalida. Intente de nuevo.")

def DD(n, matriz):
    for i in range(n):
        suma = sum(abs(matriz[i][j]) for j in range(n) if i != j)
        if abs(matriz[i][i]) < suma:
            return False
    return True

def determinante(n, matriz):
    copiamatriz = [fila[:] for fila in matriz]
    det = 1
    for i in range(n):
        for k in range(i+1, n):
            if copiamatriz[i][i] == 0:
                return 0
            factor = copiamatriz[k][i] / copiamatriz[i][i]
            for j in range(i, n):
                copiamatriz[k][j] -= factor * copiamatriz[i][j]
        det *= copiamatriz[i][i]
    return det

def jacobi(n, matriz, vector, x_inicial, max_iter, tol):
    x = x_inicial[:]
    print("\nIteracion\tVector solucion\t\tError")
    for k in range(1, max_iter+1):
        x_new = []
        for i in range(n):
            suma = sum(matriz[i][j] * x[j] for j in range(n) if i != j)
            x_new.append((vector[i] - suma) / matriz[i][i])

        error = max(abs(x_new[i] - x[i]) for i in range(n))
        x = x_new[:]
        print(f"{k}\t\t{' '.join(f'{val:.6f}' for val in x)}\t{error:.6f}")

        if error < tol:
            print(f"\nConvergencia alcanzada en {k} iteraciones.")
            print("Solucion aproximada:")
            for i, val in enumerate(x, start=1):
                print(f"x[{i}] = {val:.6f}")
            return

    print("\nNo se alcanzo la convergencia en el numero maximo de iteraciones.")

def menuPrincipal():
    print("\n\tSISTEMAS DE ECUACIONES")
    print("\tMetodos numericos - Grupo 1301\n")
    print("\tAlcantar Hernandez Jessica Esmeralda")
    print("\tFierro Ibanez Andrea Esteph")
    print("\tJaimes Molina Andrea")
    print("\tLopez Ramirez Mariana Alejandra\n")

def main():
    menuPrincipal()
    while True:
        try:
            n = int(input("Dimension de la matriz cuadrada: "))
            if n <= 0:
                raise ValueError
            break
        except ValueError:
            print("Entrada invalida para la dimension. Intente de nuevo.")

    matriz, vector = leermatriz(n)

    while True:
        mostrarmatriz(n, matriz, vector)
        try:
            corregir = int(input("\n¿Es correcta la matriz? ([SI=1], [NO=0]): "))
            if corregir not in (0, 1):
                raise ValueError
        except ValueError:
            print("Entrada invalida. Intente de nuevo.")
            continue

        if corregir == 1:
            break
        corregircoeficiente(n, matriz, vector)

    dominante = DD(n, matriz)
    det = determinante(n, matriz)

    if det != 0:
        print(f"El determinante es: {det:.6f}")

        if not dominante:
            print("La matriz NO es EDD. La convergencia no se garantiza.")

        x_inicial = []
        print("Introduce el vector inicial:")
        for i in range(n):
            while True:
                try:
                    x_inicial.append(float(input(f"x_inicial[{i+1}]: ")))
                    break
                except ValueError:
                    print("Entrada invalida. Intente de nuevo.")

        while True:
            try:
                max_iter = int(input("Maximo de iteraciones: "))
                if max_iter <= 0:
                    raise ValueError
                break
            except ValueError:
                print("Entrada invalida para el numero de iteraciones. Intente de nuevo.")

        while True:
            try:
                tol = float(input("Tolerancia: "))
                if tol <= 0:
                    raise ValueError
                break
            except ValueError:
                print("Entrada invalida para la tolerancia. Intente de nuevo.")

        jacobi(n, matriz, vector, x_inicial, max_iter, tol)
    else:
        print("El determinante es 0. El sistema no tiene solucion unica.")

    print("Programa terminado.")

if __name__ == "__main__":
    main()
```
